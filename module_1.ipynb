```python
def avPrice(n):
    x=(df['Average Price'].tail(n).sum()/n)
    return x
def profit_loss_percent(n):
    y=(df['Close Price'].iloc[494-n])
    z=(df['Close Price'].iloc[493])
    profit_loss=(z-y)
    if profit_loss <0:
        profit_loss= -(profit_loss)
    if profit_loss >0:
        profit_loss= (profit_loss)
    profit_loss_per= (profit_loss/z)*100
    return profit_loss_per
print("Average price ")
print("Avg price of 1 week is",avPrice(7))
print("Avg price of 2 week is",avPrice(14))
print("Avg price of 1 month is",avPrice(30))
print("Avg price of 3 month is",avPrice(90))
print("Avg price of 6 month is",avPrice(150))
print("Avg price of 1 year is",avPrice(365))
print(" Profit/Loss Percent")
print("Profit/Loss of 1 week is",profit_loss_percent(7))
print("Profit/Loss of 2 week is",profit_loss_percent(14))
print("Profit/Loss of 1 month is",profit_loss_percent(30))
print("Profit/Loss of 3 month is",profit_loss_percent(90))
print("Profit/Loss of 6 month is",profit_loss_percent(150))
print("Profit/Loss of 1 year is",profit_loss_percent(365))
    
    
```

```python
 #1.6
df['Day_Perc_Change'] = df['Close Price'].pct_change().fillna(0)
```

```python

df.head()
```

```python
df.insert(df.shape[1], 'Trend', 'Slight')

```

```python
# using .loc to update the value in the column
df.loc[df['Day_Perc_Change'].between(-0.005, -0.01, inclusive = True), 'Trend'] = 'Slight negative'
```

```python
    
df.loc[df['Day_Perc_Change'].between(0.005, 0.01, inclusive = True), 'Trend'] = 'Slight positive'
```

```python
df.loc[df['Day_Perc_Change'].between(-0.03, -0.01, inclusive = False),'Trend'] = 'Negative'
```

```python
df.loc[df['Day_Perc_Change'].between(0.01, 0.03, inclusive = False),'Trend'] = 'Positive'
```

```python
df.loc[df['Day_Perc_Change'].between(-0.07, -0.03, inclusive = True),'Trend'] = 'Top losers'

```

```python

df.loc[df['Day_Perc_Change'].between(0.03, 0.07, inclusive = True),'Trend'] = 'Top gainers'
```

```python
df.loc[df['Day_Perc_Change']<-0.07,'Trend']= 'Bear drop'
```

```python

df.loc[df['Day_Perc_Change']>0.07,'Trend'] = 'Bull run'
```
