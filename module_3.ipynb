```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
import datetime
```

```python
df=pd.read_csv(r'F:\ML Internship Data\GOLD.csv')
df
```

```python
df_new=df.dropna()
df_new.tail()
```

```python
x=np.array(df_new['new'])
x
```

```python
y=np.array(df_new['Pred'])
y
```

```python
x.ndim
```

```python
y.ndim
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
reg_model = LinearRegression()
reg_model.fit(x, y)
```

```python
y_predicted = reg_model.predict(x)
rmse = mean_squared_error(y, y_predicted)
r2 = r2_score(y, y_predicted)
```

```python

print('Slope:' ,reg_model.coef_)
print('Intercept:', reg_model.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)
```

```python
Gail=Gail_dt.tail(90)
Gail
```

```python
Gail_new=Gail['Close Price'].pct_change()
Nifty_new=Nifty['Close'].pct_change()
plt.figure(figsize=(15,8))
Gail_new.plot()
Nifty_new.plot()
plt.ylabel("Daily Return of Gail and NIFTY")
plt.show()
```

```python
new_data=pd.concat([Gail_dt["Close Price"], Nifty_dt["Close"]], axis=1 )
new_data=new_data.pct_change(1)
new_data.dropna(inplace=True)
new_data
```

```python

x=new_data['Close Price'].resample('M').sum()
x
```

```python
y=new_data['Close'].resample('M').sum()
y
```

```python
x1 = sm.add_constant(x)
model = sm.OLS(y, x1)
results = model.fit()
print(results.summary())
```
